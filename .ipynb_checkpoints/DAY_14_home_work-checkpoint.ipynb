#### sklearn kutubxonasi orqliy label encoding qilamiz

```python
le = LabelEncoder()

object_columns = df.select_dtypes(include=['object']).columns

label_encoder_df = df.drop(columns=object_columns)

column_value = {}

for column in object_columns:
  label_encoder_df[column] = le.fit_transform(df[column])
  column_value[column] = le.classes_

label_encoder_df.head()
```

### ma'lumotlarni qayta tiklash

```python
for column in object_columns:
  print(label_encoder_df[column].agg(lambda x: column_value[column][x]))
```

```python
factorize_df = df.copy()
for column in object_columns:
  factorize_df[column], index = pd.factorize(df[column])
```

```python
for column in object_columns:
  print(factorize_df[column].agg(lambda x: index[x]))
```

#### kerakliy funksiyalarni yaratib olamiz masalan boxplot qurish, outlierlarni uchirish va outlayerlarni almashtirish uchun qayta qayta bir xel kod yoziladi shuning uchun buni funksiya qilib olamiz. Outlierlarni uchirish va uzgartirish uchun qilingan funksiyalar funksiya qabul qiladi va usha funksiya bergan ma'lumot buyicha ishlaydi

```python
def box_plot(insurance_df, df):
  fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(15, 15))

  sns.boxplot(df['bmi'], ax=ax[0][0])
  ax[0][0].title.set_text('Without Outliers')
  sns.boxplot(insurance_df['bmi'], ax=ax[0][1])
  ax[0][1].title.set_text('With Outliers')
  ax[0][0].set_xlabel('BMI')
  ax[0][1].set_xlabel('BMI')

  sns.boxplot(df['charges'], ax=ax[1][0])
  ax[1][0].title.set_text('Without Outliers')
  sns.boxplot(insurance_df['charges'], ax=ax[1][1])
  ax[1][1].title.set_text('With Outliers')
  ax[1][0].set_xlabel('Charges')
  ax[1][1].set_xlabel('Charges')

  sns.boxplot(df['age'], ax=ax[2][0])
  ax[2][0].title.set_text('Without Outliers')
  sns.boxplot(insurance_df['age'], ax=ax[2][1])
  ax[2][1].title.set_text('With Outliers')
  ax[2][0].set_xlabel('Age')
  ax[2][1].set_xlabel('Age')
  plt.show()


def remove_outliers(df, funk):
  numeric_columns = df.select_dtypes(include=['number']).columns
  new_df = df.copy()
  for col in numeric_columns:
      lower_limit, upper_limit = funk(new_df, col)
      new_df = new_df[\
      (new_df[col] >= lower_limit) & (new_df[col] <= upper_limit)]

  return new_df




def change_outliers(df, funk):
  new_df = df.copy()
  numeric_columns = new_df.select_dtypes(include=['number']).columns
  for col in numeric_columns:
      lower_limit, upper_limit = funk(new_df, col)
      new_df[col] = new_df[col].agg(lambda x: upper_limit if x > upper_limit else lower_limit if x < lower_limit else x)

  return new_df
```

#### bu yirda tepadagu funksiyani ishlatamiz va upper_limit, va lower_limit ni hisoblovchi funksiya yaratamiz har bir yunalish uchun uziga mos funsiay yaratilga

```python
def get_outliers_gd(df, col, limit=2.6):
    mean = df[col].mean()
    std = df[col].std()
    lower_limit = mean - limit * std
    upper_limit = mean + limit * std
    return lower_limit, upper_limit

gd_df_1 = remove_outliers(df, get_outliers_gd)

box_plot(df, gd_df_1)
print(shape[0] - gd_df_1.shape[0])
```

#### Outlierlarni min_limit va max_limitga tenglash

```python
gd_df_2 = change_outliers(df, get_outliers_gd)
print(shape[0] - gd_df_2.shape[0])
box_plot(df, gd_df_2)
```

```python
def get_outliers_iqr(df, col, limit=1.5):
  q1 = df[col].quantile(.25)
  q3 = df[col].quantile(.75)
  iqr = q3 - q1
  lower_limit = q1 - limit * iqr
  upper_limit = q3 + limit * iqr
  return lower_limit, upper_limit


iqr_df_1 = remove_outliers(df, get_outliers_iqr)
print(shape[0] - iqr_df_1.shape[0])
box_plot(df, iqr_df_1)
```

```python
iqr_df_2 = change_outliers(df, get_outliers_iqr)
print(shape[0] - iqr_df_2.shape[0])
box_plot(df, iqr_df_2)
```

```python
def get_outliers_quartile(df, col):
  new_df = df.copy()
  new_df.sort_values(col, inplace=True)
  lower_limit = new_df[col].quantile(.05)
  upper_limit = new_df[col].quantile(.95)
  return lower_limit, upper_limit
```

```python
quartile_df_1 = remove_outliers(df, get_outliers_quartile)
print(shape[0] - quartile_df_1.shape[0])
box_plot(df, quartile_df_1)
```

```python
quartile_df_2 = change_outliers(df, get_outliers_quartile)
print(shape[0] - quartile_df_2.shape[0])
box_plot(df, quartile_df_2)
```
