```python
import pandas as pd
import numpy as np
```

```python
internet_usage_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/Percentage%20of%20individuals%20using%20the%20internet.csv')
freshwater_withdrawals_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/Uzbekistan_Freshwater_Withdrawals_1991_2023.csv',
                                        skiprows=2)
births_2010_2023_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/uzbekistan_births_2010_2023.csv')
life_expectancy_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/Uzbekistan_Life_Expectancy__1991-2023_.csv')
unemployment_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/Uzbekistan_Unemployment_1991_2023.csv')
national_debt_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/UZ_National%20Debt%20(USD).csv')
deaths_2010_2023_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/uzbekistan_deaths_2010_2023.csv')
economic_data_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/Uzbekistan_Economic_Pivoted_1991_2023.csv',
                               skiprows=23)
population_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/UZ_Population.csv')
forest_area_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/Uzbekistan_Forest_Area_1991_2023.csv',
                             skiprows=1)
regions_deaths_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/uzbekistan_regions_deaths_2010_2023.csv')
net_migration_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/Uzbekistan_Net_Migration_1991_2023.csv')
gnp_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/UZ_GNP.csv',
                     skiprows=2)
birth_rate_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/uzbekistan_birth_rate_1991_2023.csv')
inflation_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/Uzbekistan_Inflation_1991_2023.csv')
regions_births_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/uzbekistan_regions_births_2010_2023.csv')
gdp_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/UZ_GDP.csv')
death_rate_df = pd.read_csv('https://codernet.uz/data/stat-uz/datasets/uzbekistan_death_rate_1991_2023.csv')
```

```python
datasets = [internet_usage_df,
            freshwater_withdrawals_df,
            births_2010_2023_df,
            life_expectancy_df,
            unemployment_df,
            national_debt_df,
            deaths_2010_2023_df,
            economic_data_df,
            population_df,
            forest_area_df,
            regions_deaths_df,
            net_migration_df,
            gnp_df,
            birth_rate_df,
            inflation_df,
            regions_births_df,
            gdp_df,
            death_rate_df]
```

```python
main_df = main_df.merge(internet_usage_df, on='Year', how='left')
main_df = main_df.merge(freshwater_withdrawals_df, on='Year', how='left')
main_df = main_df.merge(births_2010_2023_df, on='Year', how='left')
main_df = main_df.merge(life_expectancy_df, on='Year', how='left')
main_df = main_df.merge(unemployment_df, on='Year', how='left')
main_df = main_df.merge(national_debt_df, on='Year', how='left')
main_df = main_df.merge(deaths_2010_2023_df, on='Year', how='left')
main_df = main_df.merge(population_df, on='Year', how='left')
main_df = main_df.merge(forest_area_df, on='Year', how='left')
main_df = main_df.merge(net_migration_df, on='Year', how='left')
main_df = main_df.merge(gnp_df, on='Year', how='left')
main_df = main_df.merge(birth_rate_df, on='Year', how='left')
main_df = main_df.merge(inflation_df, on='Year', how='left')
main_df = main_df.merge(gdp_df, on='Year', how='left')
main_df = main_df.merge(death_rate_df, on='Year', how='left')
main_df = main_df.merge(economic_data_df, on='Year', how='left')
main_df = main_df.merge(regions_deaths_df, on='Year', how='left')
main_df = main_df.merge(regions_births_df, on='Year', how='left')

main_df.columns
```
