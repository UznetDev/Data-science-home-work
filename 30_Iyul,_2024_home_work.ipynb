```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import cross_val_score, KFold, train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = fetch_california_housing()
X, y = data.data, data.target
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Predict on the testing data
y_pred = model.predict(X_test)

# Compute the RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Print the RMSE
print("RMSE: ", rmse)
```

```python
data = fetch_california_housing()
X, y = data.data, data.target

# Initialize the Linear Regression model
model = LinearRegression()

# Define the cross-validation strategy (5-fold)
kf = KFold(n_splits=10, shuffle=True, random_state=0)

# Perform cross-validation and compute RMSE for each fold
rmse_scores = np.sqrt(-cross_val_score(model, X, y, cv=kf, scoring='neg_mean_squared_error'))

# Print the RMSE scores and the mean RMSE
print("RMSE scores for each fold: ", rmse_scores)
print("Mean RMSE: ", np.mean(rmse_scores))
```

```python
# Load the California Housing dataset
data = fetch_california_housing()
X, y = data.data, data.target

# Define the range of alpha and l1_ratio values to loop over
alphas = [0.1, 1.0, 10.0, 100, 1000, 10000]
l1_ratios = [0, 0.1, 0.5, 0.9]

# Define the cross-validation strategy (5-fold)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize a list to store results
results = []

# Loop over the different combinations of alpha and l1_ratio
for alpha in alphas:
    for l1_ratio in l1_ratios:
        # Initialize the ElasticNet model with the current alpha and l1_ratio
        elasticnet_model = ElasticNet(alpha=alpha, l1_ratio=l1_ratio)

        # Perform cross-validation and compute RMSE for each fold
        rmse_scores = np.sqrt(-cross_val_score(elasticnet_model, X, y, cv=kf, scoring='neg_mean_squared_error'))

        # Store the results in the list
        results.append({
            'alpha': alpha,
            'l1_ratio': l1_ratio,
            'mean_rmse': np.mean(rmse_scores),
            'std_rmse': np.std(rmse_scores),
            'rmse_scores': rmse_scores
        })

# Convert the results list to a DataFrame
results_df = pd.DataFrame(results)

# Print the DataFrame
results_df
```

```python
# Ma'lumotlarni yuklash
data = fetch_california_housing()
X, y = data.data, data.target

# Parametrlar
degrees = [1, 2, 3]  # Polynomial darajalari
alphas = [0.01, 0.1, 1.0, 10.0]  # Ridge va Lasso uchun alpha
l1_ratios = [0.1, 0.5, 0.9]  # ElasticNet uchun l1_ratio

# Natijalarni saqlash uchun bo'sh ro'yxat
results = []

# Cross-validation tayyorlash
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Modelni optimizatsiya qilish
for degree in degrees:
    # Polynomial features yaratish
    poly = PolynomialFeatures(degree)
    X_poly = poly.fit_transform(X)

    # Ridge modelini sinash
    for alpha in alphas:
        ridge = Ridge(alpha=alpha)
        scores = cross_val_score(ridge, X_poly, y, scoring='neg_mean_squared_error', cv=kf)
        rmse = np.sqrt(-scores.mean())
        mae = -cross_val_score(ridge, X_poly, y, scoring='neg_mean_absolute_error', cv=kf).mean()
        results.append(('Ridge', degree, alpha, None, rmse, mae))

    # Lasso modelini sinash
    for alpha in alphas:
        lasso = Lasso(alpha=alpha)
        scores = cross_val_score(lasso, X_poly, y, scoring='neg_mean_squared_error', cv=kf)
        rmse = np.sqrt(-scores.mean())
        mae = -cross_val_score(lasso, X_poly, y, scoring='neg_mean_absolute_error', cv=kf).mean()
        results.append(('Lasso', degree, alpha, None, rmse, mae))

    # ElasticNet modelini sinash
    for alpha in alphas:
        for l1_ratio in l1_ratios:
            elastic_net = ElasticNet(alpha=alpha, l1_ratio=l1_ratio)
            scores = cross_val_score(elastic_net, X_poly, y, scoring='neg_mean_squared_error', cv=kf)
            rmse = np.sqrt(-scores.mean())
            mae = -cross_val_score(elastic_net, X_poly, y, scoring='neg_mean_absolute_error', cv=kf).mean()
            results.append(('ElasticNet', degree, alpha, l1_ratio, rmse, mae))
```

```python
# Eng yaxshi natijani topish
best_model = min(results, key=lambda x: x[4])
print(f"Eng yaxshi model: {best_model[0]}")
print(f"Polynomial darajasi: {best_model[1]}")
print(f"Alpha: {best_model[2]}")
if best_model[0] == 'ElasticNet':
    print(f"L1_ratio: {best_model[3]}")
print(f"RMSE: {best_model[4]}")
print(f"MAE: {best_model[5]}")
```
