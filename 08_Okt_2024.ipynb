```python
df['Pilot Name'] = df['Pilot Name'].drop_duplicates()
```

```python
df = df.drop_duplicates(subset=['Passenger ID'])
```

```python
df['Flight Status'].unique()
```

```python
df['target'] = df['Flight Status'].agg(lambda x: 1 if x == 'Cancelled' else 0)
```

```python
df['target'].value_counts()
```

```python
new_df = df.drop(['First Name', 'Last Name', 'Pilot Name'], axis=1)


new_df['Departure Date'] = new_df['Departure Date'].str.replace('-', '/')
new_df['Departure Date'] = pd.to_datetime(new_df['Departure Date'], errors='coerce')
new_df['weekday'] = new_df['Departure Date'].dt.weekday
new_df['day'] = new_df['Departure Date'].dt.day
new_df['month'] = new_df['Departure Date'].dt.month
new_df['year'] = new_df['Departure Date'].dt.year
```

```python
my_df = new_df.groupby('Gender')['target'].value_counts().unstack()
```

```python
my_df
```

```python
my_df['Persentage event'] = my_df[0] / sum(my_df[0])
```

```python
my_df['Persentage non event'] = my_df[1] / sum(my_df[1])
```

```python
my_df['WOE'] = np.log(my_df['Persentage event'] / my_df['Persentage non event'])
```
