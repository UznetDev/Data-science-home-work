#### NaN qiymatlarni uzgartirish uchun funksiay tuzin olamiz

```python
def fillna(data, group_cols, col, funktion=np.median, value=False):
    return (
        data
        .groupby(by=group_cols)[col]
        .transform(lambda x: value if value else x.fillna(funktion(x.dropna())))
    )
```

#### NaN qiymatlari bor ustunlarni eng mos ustunlarning qiymatiga nisbatann almashtirishga harakat qilib kuramiz

```python
# SAT_AVG_ALL
df['SAT_AVG_ALL'] = fillna(df, ['Degree_Type','REGION'], "SAT_AVG_ALL", np.median)
df['SAT_AVG_ALL'] = fillna(df, ['INSTNM','REGION'], "SAT_AVG_ALL", np.median)
df['SAT_AVG_ALL'] = fillna(df, ['Degree_Type','INSTNM'], "SAT_AVG_ALL", np.median)
df['SAT_AVG_ALL'] = fillna(df, ['Degree_Type'], "SAT_AVG_ALL", np.median)
```

```python
# COMPL_RPY_5YR_RT
df["COMPL_RPY_5YR_RT"] = pd.to_numeric(df["COMPL_RPY_5YR_RT"], errors='coerce')
df["COMPL_RPY_5YR_RT"] = pd.to_numeric(df["COMPL_RPY_5YR_RT"], errors='coerce')
df['COMPL_RPY_5YR_RT'] = fillna(df, ['REGION', 'Ownership', "Degree_Type", "Locale_Short"], "COMPL_RPY_5YR_RT")
df['COMPL_RPY_5YR_RT'] = fillna(df, ['REGION', 'Ownership'], "COMPL_RPY_5YR_RT")
```

```python
# ADM_RATE_ALL
df['ADM_RATE_ALL'] = fillna(df, ['Degree_Type','REGION'], "ADM_RATE_ALL")
df['ADM_RATE_ALL'] = fillna(df, ['Degree_Type'], "ADM_RATE_ALL")
```

```python
# PCTPELL
df['PCTPELL'] = fillna(df, ['Degree_Type','REGION'], "PCTPELL")
df['PCTPELL'] = fillna(df, ['REGION'], "PCTPELL")
df['PCTPELL'] = fillna(df, ['Degree_Type'], "PCTPELL")
```

```python
# UG
df['UG'] = fillna(df, ['REGION', 'INSTNM'], "UG", np.mean)
df['UG'] = fillna(df, ['Degree_Type', 'INSTNM'], "UG", np.mean)
df['UG'] = fillna(df, ['Degree_Type', 'REGION'], "UG", np.mean)
df['UG'] = fillna(df, ['Degree_Type'], "UG", np.mean)
```

```python
# AVGFACSAL
df['AVGFACSAL'] = fillna(df, ['REGION', 'INSTNM'], "AVGFACSAL", np.median)
df['AVGFACSAL'] = fillna(df, ['REGION', 'Tuition'], "AVGFACSAL", np.median)
df['AVGFACSAL'] = fillna(df, ['Ownership', 'Degree_Type'], "AVGFACSAL", np.median)
df['AVGFACSAL'] = fillna(df, ['Degree_Type'], "AVGFACSAL", np.median)
```

```python
# PCTFLOAN
df['PCTFLOAN'] = fillna(df, ['REGION', 'Regions'], "PCTFLOAN")
```

```python
# LOCALE
df['LOCALE'] = fillna(df, ['REGION', 'Regions', 'INSTNM'], "LOCALE")
df['LOCALE'] = fillna(df, ['REGION', 'Regions'], "LOCALE")
```

```python
# Tuition
df['Tuition'] = fillna(df, ['Degree_Type', 'AVGFACSAL'], "Tuition", np.mean)
df['Tuition'] = fillna(df, ['Degree_Type', 'REGION'], "Tuition", np.mean)
```

```python
# Locales
df['Locales'] = df['Locales'].fillna(df['Locales'].mode()[0])
```

```python
# Locale_Short
df['Locale_Short'] = df['Locale_Short'].fillna(df['Locale_Short'].mode()[0])
```

```python
# DEBT_MDN
df['DEBT_MDN'] = pd.to_numeric(df['DEBT_MDN'], errors='coerce')
df['DEBT_MDN'] = fillna(df, ['Locales', 'REGION'], "DEBT_MDN")
df['DEBT_MDN'] = fillna(df, ['REGION'], "DEBT_MDN")
```

```python
# MEDIAN_HH_INC
df['MEDIAN_HH_INC'] = pd.to_numeric(df['MEDIAN_HH_INC'], errors='coerce')
df['MEDIAN_HH_INC'] = fillna(df, ['REGION', 'Locales'], "MEDIAN_HH_INC")
```

```python
# WOMENONLY
df['WOMENONLY'] = fillna(df, ['Degree_Type'], "WOMENONLY")
```

```python
# MENONLY
df['MENONLY'] = fillna(df, ['Degree_Type'], "MENONLY")
```
