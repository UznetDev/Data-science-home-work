#### kerakliy kutubxonalrni chaqirib olamiz

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Dataframe mizni uqibolamiz

```python
df = pd.read_csv('Final.csv')
```

```python
df.info()
```

##### Dataframe ichidan numeric ustunlarni yangi DataFramega utkazib olamiz

```python
object_column = df.select_dtypes(include='object').columns
num_df = df.drop(columns=object_column)
num_df.info()
```

#### Kid plot uchun funksiya tuzib olamiz va numeric datframemizda sinab kuramziz

```python
def histogram(df: pd.DataFrame):

    count = len(df.columns)
    fig, ax = plt.subplots(nrows=(count + 1) // 2, ncols=2, figsize=(12, count * 2))
    if count == 1:
        ax = [ax]
    else:
        ax = ax.flatten()

    for i in range(len(df.columns)):
        col = df.columns[i]
        sns.kdeplot(df[col], ax=ax[i], fill=True)
        ax[i].set_title(col)
    plt.show()
histogram(num_df)
```

#### Standardisation

```python
standardisation_df = (num_df - num_df.mean()) / num_df.std()
```

```python
histogram(standardisation_df)
```

#### Min-Max Scaling

```python
min_max_scaled_df = (num_df - num_df.min()) / (num_df.max() - num_df.min())
histogram(min_max_scaled_df)
```

#### Mean Normalisation

```python
mean_narmalised_df = (num_df - num_df.mean()) / (num_df.max() - num_df.min())
histogram(mean_narmalised_df)
```

#### MaxAbsScaling

```python
max_abs_scaling_df = num_df / num_df.abs().max()
histogram(max_abs_scaling_df)
```

#### Scaling to median and IQR

```python
robus_scaling_df = (num_df - num_df.median()) / (num_df.quantile(0.75) - num_df.quantile(0.25))
histogram(robus_scaling_df)
```
