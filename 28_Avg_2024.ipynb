### Class work

```python
%%sql SELECT date,
      (home_team_goal + away_team_goal) AS goals,
      RANK() OVER(ORDER BY home_team_goal + away_team_goal) AS rank
FROM match
WHERE season='2008/2009';
```

```python
%%sql SELECT DISTINCT season FROM match;
```

```python
%%sql SELECT date,
      (home_team_goal + away_team_goal) AS goals,
      RANK() OVER(ORDER BY home_team_goal + away_team_goal DESC) AS rank
FROM match
WHERE season='2008/2009';
```

```python
%%sql SELECT date,
      (home_team_goal + away_team_goal) AS goals,
      RANK() OVER(ORDER BY home_team_goal + away_team_goal DESC) AS rank
FROM match
WHERE season IS NOT NULL;
```

```python
%%sql SELECT
      date,
      (home_team_goal + away_team_goal) AS goals,
      AVG(home_team_goal + away_team_goal) OVER(PARTITION BY season) AS rank
FROM match;
```

```python
%%sql
SELECT
      c.name,
      m.season,
      date,
      (home_team_goal + away_team_goal) AS goals,
      SUM(home_team_goal + away_team_goal) OVER(PARTITION BY m.season, c.name) AS sum_goal
FROM country AS c
LEFT JOIN match AS m
ON c.id = m.country_id
WHERE name!='Belgium';
```

```python
%%sql
SELECT date,
       home_team_goal,
       away_team_goal,
       SUM(home_team_goal) OVER(ORDER BY date
                                ROWS BETWEEN UNBOUNDED PRECEDING
                                AND CURRENT ROW) AS cum_sum
FROM match
WHERE home_team_api_id=10000 AND season IS NOT NULL;
```

```python
%%sql
SELECT date,
       home_team_goal,
       away_team_goal,
       AVG(home_team_goal) OVER(ORDER BY date
                                ROWS BETWEEN UNBOUNDED PRECEDING
                                AND CURRENT ROW) AS cum_sum
FROM match
WHERE home_team_api_id=10000 AND season IS NOT NULL;
```

```python
%%sql
SELECT date,
       home_team_goal,
       away_team_goal,
       AVG(home_team_goal) OVER(ORDER BY date
                                ROWS BETWEEN UNBOUNDED PRECEDING
                                AND CURRENT ROW) AS cum_sum,
       (SELECT AVG(home_team_goal)
       FROM match) AS avg_goals
FROM match
WHERE home_team_api_id=10000 AND season IS NOT NULL;
```
