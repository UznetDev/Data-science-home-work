### 2024 yilda ma ºlumotlarni ishlab chiquvchilarining ish haqi dataseti ustida ishlaymiz

**experience_level:** Kasbiy tajriba darajasi (masalan, yosh, o'rta, katta).

**employment_type:** Ish shartnoma turi (masalan, to'liq ish vaqti, qismiy ish, shartnoma).

**job_title:** Xodimning ma'lum bir lavozimi yoki unvoni (masalan, Ma'lumotlar Muhandisi).

**salary:** Qabul qilingan mablag', asl valyutada.

**salary_currency:** Qabul qilingan valyuta.

**salary_in_usd:** Taqqoslash maqsadida AQSh dollari konvertatsiyasi.

**employee_residence:** Xodimning yashash joyi.

**remote_ratio:** Masofaviy ishning qancha foizini bajarish mumkin.

**company_location:** Kompaniya joylashuvi, geografik joylashuvi.

**company_size:** Odatda xodimlar soni asosida hisoblanadi.

#### UI yaratish

```python
import matplotlib.pyplot as plt
import seaborn as sns

average_salary = df.groupby('experience_level')['salary_in_usd'].mean().reset_index()
plt.figure(figsize=(30, 15))

sns.barplot(y='salary_in_usd', x='experience_level', data=average_salary)

plt.xticks(fontsize=32, rotation=90)
plt.yticks(fontsize=32) 

plt.title('Darajalar buyicha ish haqqi!', fontsize=45)
plt.xlabel("O'rtacha ish haqi (USD)", fontsize=42)
plt.ylabel('Ish haqqi', fontsize=42)
plt.show()
```
