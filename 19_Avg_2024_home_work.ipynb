```python
import pandas as pd
```

```python
# Read datasets from CSV files
films = pd.read_csv('films.csv', header=None)
films = films.rename(columns={0: 'id',
                              1: 'title',
                              2: 'releas_date',
                              3: 'country',
                              4: 'duratioon',
                              5: 'language',
                              6: 'certificatipon',
                              7: 'gross',
                              8: 'budget'})
reviews = pd.read_csv('reviews.csv', header=None)
reviews = reviews.rename(columns={0: 'id',
                                  1: 'film_id',
                                  2: 'num_user',
                                  3: 'num_critic',
                                  4: 'imdb_score',
                                  5: 'num_votes',
                                  6: 'facebook_likes'})
people = pd.read_csv('people.csv', header=None)
people = people.rename(columns={0: 'id',
                                1: 'name',
                                2: 'birthdate',
                                3: 'deathdate'})
roles = pd.read_csv('roles.csv', header=None)
roles = roles.rename(columns={0: 'id',
                              1: 'film_id',
                              2: 'person_id',
                              3: 'role'})

# Drop tables if they exist, then persist the DataFrames to the SQL database
%sql DROP TABLE IF EXISTS films;
%sql --persist films

%sql DROP TABLE IF EXISTS reviews;
%sql --persist reviews

%sql DROP TABLE IF EXISTS people;
%sql --persist people

%sql DROP TABLE IF EXISTS roles;
%sql --persist roles
```

Urtachadan kuproq foyda keltira olgan filmlar

```python
%%sql SELECT *
FROM films
WHERE gross > (SELECT AVG(gross) FROM films) AND gross > budget;
```

Kinolaridan zarar kurish buyicha mamlakatlar riytingi

```python
%%sql
SELECT fm.country,
       COUNT(*) AS loss_films_count,
       SUM(fm.gross - fm.budget) AS loss_sum
FROM films AS fm
WHERE fm.gross < fm.budget
GROUP BY fm.country
HAVING loss_films_count > 0
ORDER BY loss_films_count DESC;
```

Kinolari zarar kurgan aktyorlar riytingi

```python
%%sql
SELECT people.id, people.name,
       COUNT(*) AS count_loss,
       SUM(fm.gross - fm.budget) AS total_loss
FROM people
INNER JOIN roles AS rl ON people.id = rl.person_id
INNER JOIN films AS fm ON rl.film_id = fm.id
WHERE fm.gross < fm.budget
GROUP BY people.id, people.name
HAVING COUNT(*) > 0
ORDER BY count_loss DESC;
```

Budjedga nisbatan eng kup foyda olgan filmlar

```python
%%sql
SELECT *,
       ((gross - budget) / budget) * 100 AS `profit %`
FROM films
WHERE budget < gross
ORDER BY `profit %` DESC;
```
